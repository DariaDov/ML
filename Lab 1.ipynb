```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
TRAINING_FILE_NAME = "lab_1_train.csv"
TEST_FILE_NAME = "lab_1_test.csv"
```

```python
with open(TRAINING_FILE_NAME, 'r') as f:
    train_data = csv.reader(f)
    
    x = []
    y = []
    
    next(train_data)
    
    for row in train_data:
        x.append(float(row[1]))
        y.append(float(row[2]))
    
#     print(x)
#     print(y)
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.title('Train dataset')

plt.scatter(x, y)
plt.plot()
plt.show()
```

```python
def cost_function(b0, b1, x, y):
    j = 0
    n = len(x)
    for i in range(n):
        j += (y[i] - (b0 + b1 * x[i])) ** 2
    return j/n
```

```python
#Gradient Descent

def gradient_descent():
    a = 0.01 # альфа
    epochs = 10000 # кількість епох

    b0 = 0
    b1 = 0

    n = len(x)

    for i in range(epochs):
        print("Epoch number", i, ":")
        print("b0 =", b0)
        print("b1 =", b1)
        print("Cost function =", cost_function(b0, b1, x, y))
        print("")
        
        g_b0 = 0
        g_b1 = 0
        
        for i in range(n):    
            y_p = b0 + b1 * x[i]
            
            g_b0 += (y[i] - y_p) 
            g_b1 += x[i] * (y[i] - y_p) 
            
        b0 -= a * (-2/n) * g_b0
        b1 -= a * (-2/n) * g_b1
        
        
    print("Epoch number", epochs, ":")
    print("b0 =", b0)
    print("b1 =", b1)
    print("Cost function =", cost_function(b0, b1, x, y))
    print("")
        
    return b0, b1
```

```python
b0, b1 = gradient_descent()
```

```python
print("b0 = ", b0)
print("b1 = ", b1)
```

```python
with open(TEST_FILE_NAME, 'r') as f_test:
    test_data = csv.reader(f_test)
    
    x_test = []
    y_test = []
    
    next(test_data)
    
    for row in test_data:
        x_test.append(float(row[1]))
        y_test.append(float(row[2]))
    
#     print(x_test)
#     print(y_test)
```

```python
test_c_f = cost_function(b0, b1, x_test, y_test)
print("Cost function for test dataset: ", test_c_f)
```

```python
p_y = [b0 + b1 * i for i in x + x_test]

plt.xlabel('x')
plt.ylabel('y')

plt.plot(x, y, 'o')
plt.plot(x_test, y_test, 'o')
plt.plot(x + x_test, p_y)

plt.legend(["Train dataset", "Test dataset", "Linear regression"])
plt.show()
```
